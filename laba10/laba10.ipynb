## Завдання 1

```python
def ispalindrom(text):
    text = text.replace(" ", "").lower()
    if text == text[::-1]:
        print("It's a palindrom")
    else:
        print("It's not a palindrom")

text = input("Введіть текст:")
ispalindrom(text)

text = input("Введіть текст:")
ispalindrom(text)
```

## Завдання 3

```python
def calculate_life_number(date):
    digits = ''.join(filter(str.isdigit, date))
    
    if not digits:
        return "Помилка: введіть хоча б одну цифру."

    digit_sum = sum(int(d) for d in digits)

    while digit_sum > 9:
        digit_sum = sum(int(d) for d in str(digit_sum))

    return digit_sum

if __name__ == "__main__":
    user_date = input("Введіть дату народження (наприклад, 19991229): ")
    
    result = calculate_life_number(user_date)
    
    print(f"Ваша Цифра життя: {result}")
```

## Завдання 1.1

```python
def find_word_in_sequence():
    word = input("Введіть слово (що шукаємо): ")
    sequence = input("Введіть комбінацію символів (де шукаємо): ")
    
    word = word.lower()
    sequence = sequence.lower()

    found = True
    start_index = 0 

    for char in word:
        index = sequence.find(char, start_index)

        if index == -1:
            found = False
            break

        start_index = index + 1

    if found:
        print("Yes")
    else:
        print("No")

if __name__ == "__main__":
    find_word_in_sequence()
```

## Завдання 2.2

```python
def calculate_life_number(date):
    digits = ''.join(filter(str.isdigit, date))
    
    if not digits:
        raise ValueError("Введені дані не містять цифр.")

    digit_sum = sum(int(d) for d in digits)

    while digit_sum > 9:
        digit_sum = sum(int(d) for d in str(digit_sum))

    return digit_sum

if __name__ == "__main__":
    
    while True:
        try:
            date_input = input("Введіть дату народження (або 'exit' для виходу): ")
            
            if date_input.lower() == 'exit':
                print("До побачення!")
                break
            
            life_number = calculate_life_number(date_input)
            
            print(f"Ваша Цифра життя: {life_number}")
            break 
            
        except ValueError as e:
            print(f"Помилка: {e} Будь ласка, спробуйте ще раз.\n")
        except Exception as e:
            print(f"Виникла несподівана помилка: {e}. Спробуйте ще раз.\n")
```

## Завдання 3.3

```python
def read_int(prompt, min, max):
    while True:
        try:
            user_input = input(prompt)

            value = int(user_input)
            
            if min <= value <= max:
                return value
            else:
                print(f"Error: the value is not within permitted range ({min}..{max})")
                
        except ValueError:
            print("Error: wrong input")

if __name__ == "__main__":
    v = read_int("Введіть число від -10 до 10: ", -10, 10)
    print("Число:", v)
```
