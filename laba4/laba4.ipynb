```python
income = float(input("Enter the annual income: "))

tax_threshold = 85528.0

if income <= tax_threshold:
    tax = (income * 0.18) - 556.02
else:
    surplus = income - tax_threshold
    tax = 14839.02 + (surplus * 0.32)

if tax < 0:
    tax = 0.0

tax = round(tax, 0)
print("The tax is:", tax, "thalers")
```

## Завдання 5

```python
year = int(input("Enter a year: "))

if year < 1582:
    print("Not within the Gregorian calendar period")
else:
    if year % 4 != 0:
        print("Common year")
    elif year % 100 != 0:
        print("Leap year")
    elif year % 400 != 0:
        print("Common year")
    else:
        print("Leap year")
```

## Завдання 6

```python
secret_number = 777

print(
"""
+================================+
| Welcome to my game, muggle!    |
| Enter an integer number        |
| and guess what number I've     |
| picked for you.                |
| So, what is the secret number? |
+================================+
""")

while True:
    user_guess = int(input("Введіть секретне число: "))

    if user_guess == secret_number:
        break
    else:
        print("Ха-ха! Ви застрягли у моїй петлі!")

print(f"Ви вгадали! Число було {secret_number}.")
print("Молодець, магле! Тепер ти вільний.")
```

```python
user_word = input("Введіть слово: ")
user_word = user_word.upper()

for letter in user_word:
    if letter == 'A':
        continue
    elif letter == 'E':
        continue
    elif letter == 'I':
        continue
    elif letter == 'O':
        continue
    elif letter == 'U':
        continue
    else:
        print(letter)
```

## Завдання 10

```python
word_without_vowels = ""

user_word = input("Будь ласка, введіть слово: ")

user_word = user_word.upper()

for letter in user_word:
    if letter == "A" or letter == "E" or letter == "I" or letter == "O" or letter == "U":
        continue
    else:
        word_without_vowels += letter

print(word_without_vowels)
```

## Завдання 11

```python
blocks = int(input("Введіть кількість блоків: "))

height = 0
blocks_in_layer = 1

while blocks >= blocks_in_layer:
    blocks -= blocks_in_layer
    
    height += 1
    
    blocks_in_layer += 1

print("The height of the pyramid:", height)
```

## Завдання 12

```python
c0 = int(input("Введіть будь-яке невід’ємне та ненульове ціле число: "))

steps = 0

if c0 > 0:
    while c0 != 1:
        if c0 % 2 == 0:
            c0 = c0 // 2
        else:
            c0 = 3 * c0 + 1
        
        print(c0)
        
        steps += 1
    
    print("steps =", steps)
else:
    print("Будь ласка, введіть число більше за нуль.")
```
