## Завдання 1

```python
def mysplit(strng):
    if strng == '' or strng.isspace():
        return []
    
    list_of_words = []
    word = ""
    
    for char in strng:
        if char.isspace():
            if word != "":
                list_of_words.append(word)
                word = ""
        else:
            word += char
            
    if word != "":
        list_of_words.append(word)
        
    return list_of_words

print(mysplit("To be or not to be, that is the question"))
print(mysplit("To be or not to be,that is the question"))
print(mysplit("   "))
print(mysplit(" abc "))
print(mysplit(""))
```

## Завдання 3

```python
def caesar_ua():
    alphabet = "АБВГҐДЕЄЖЗИІЇЙКЛМНОПРСТУФХЦЧШЩЬЮЯ"
    
    text = input("Введіть повідомлення: ").upper()
    shift = 1
    
    cipher_text = ""
    
    for char in text:
        if char in alphabet:
            original_index = alphabet.find(char)
            new_index = (original_index + shift) % len(alphabet)
            
            cipher_text += alphabet[new_index]
        else:
            continue

    print(f"Зашифроване: {cipher_text}")

caesar_ua()
```

## Завдання 4

```python
def cipher_decrypt():
    alphabet = "АБВГҐДЕЄЖЗИІЇЙКЛМНОПРСТУФХЦЧШЩЬЮЯ"
    
    cryptogram = input("Enter your cryptogram: ").upper()
    
    decoded_text = ""
    
    for char in cryptogram:
        if char in alphabet:
            index = alphabet.find(char)

            new_index = (index - 1) % len(alphabet)
            
            decoded_text += alphabet[new_index]
        else:
            decoded_text += char
            
    print(decoded_text)

if __name__ == "__main__":
    cipher_decrypt()
```

## Завдання 1.1

```python
def caesar_cipher():
    text = input("Введіть рядок для шифрування: ")

    shift = 0
    while True:
        try:
            user_input = input("Введіть зсув (ціле число від 1 до 25): ")
            shift = int(user_input)
            
            if 1 <= shift <= 25:
                break  
            else:
                print("Помилка! Число має бути в діапазоні від 1 до 25.")
        except ValueError:
            print("Помилка! Ви ввели не число. Спробуйте ще раз.")

    cipher_text = ""

    for char in text:
        if char.isalpha():
            if char.isupper():
                base = ord('A')
            else:
                base = ord('a')
            
            new_code = base + (ord(char) - base + shift) % 26
            cipher_text += chr(new_code)
        else:
            cipher_text += char

    print("Закодований текст:", cipher_text)

if __name__ == "__main__":
    caesar_cipher()
```
