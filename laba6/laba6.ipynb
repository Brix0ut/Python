## Завдання 3

```python
def is_year_leap(year):
    """
    Перевіряє, чи є рік високосним.
    """
    if (year % 4 == 0 and year % 100 != 0) or (year % 400 == 0):
        return True
    else:
        return False

def days_in_month(year, month):
    """
    Повертає кількість днів у вказаному місяці та році.
    """
    if month < 1 or month > 12:
        return None
    
    month_days = [0, 31, 28, 31, 30, 31, 30, 31, 31, 30, 31, 30, 31]
    
    if is_year_leap(year) and month == 2:
        return 29
    
    return month_days[month]

def day_of_year(year, month, day):
    """
    Повертає порядковий номер дня в році.
    """
    max_days_in_current_month = days_in_month(year, month)
    
    if max_days_in_current_month is None:
        return None
        
    if day < 1 or day > max_days_in_current_month:
        return None
        
    total_days = 0
    for m in range(1, month):
        total_days += days_in_month(year, m)
        
    total_days += day
    
    return total_days

print("--- Запуск тестів ---")

print("31.12.2000 (високосний) ->", day_of_year(2000, 12, 31)) 

print("31.12.1999 (звичайний) ->", day_of_year(1999, 12, 31)) 
print("01.01.2000 (високосний) ->", day_of_year(2000, 1, 1))
print("01.01.1999 (звичайний) ->", day_of_year(1999, 1, 1))  
print("01.03.2000 (високосний) ->", day_of_year(2000, 3, 1))
print("01.03.2001 (звичайний) ->", day_of_year(2001, 3, 1))  

print("32.01.2001 (невірний день) ->", day_of_year(2001, 1, 32)) 
print("29.02.2001 (невірний день) ->", day_of_year(2001, 2, 29)) 
print("01.13.2001 (невірний місяць) ->", day_of_year(2001, 13, 1))
print("01.00.2001 (невірний місяць) ->", day_of_year(2001, 0, 1))  
```

```python
def is_prime(num):
    """
    Перевіряє, чи є число простим.
    """
    if num <= 1:
        return False    
    for i in range(2, int(num**0.5) + 1):
        if num % i == 0:
            return False
            
    return True

for i in range(1, 20):
    if is_prime(i + 1):
            print(i + 1, end=" ")
```

## Завдання 5

```python
METERS_IN_MILE = 1609.344
LITERS_IN_GALLON = 3.785411784
METERS_IN_100KM = 100000

def liters_100km_to_miles_gallon(liters):
    """
    Конвертує витрату палива з л/100км у милі/галон.
    """
    
    miles_in_100km = METERS_IN_100KM / METERS_IN_MILE
    
    gallons_in_liters = liters / LITERS_IN_GALLON
    
    miles_per_gallon = miles_in_100km / gallons_in_liters
    
    return miles_per_gallon

def miles_gallon_to_liters_100km(miles):
    """
    Конвертує витрату палива з милі/галон у л/100км.
    """
    
    meters_travelled = miles * METERS_IN_MILE

    meters_per_liter = meters_travelled / LITERS_IN_GALLON
    
    liters_per_100km = METERS_IN_100KM / meters_per_liter
    
    return liters_per_100km

print(liters_100km_to_miles_gallon(3.9))
print(liters_100km_to_miles_gallon(7.5))
print(liters_100km_to_miles_gallon(10.))
print(miles_gallon_to_liters_100km(60.3))
print(miles_gallon_to_liters_100km(31.4))
print(miles_gallon_to_liters_100km(23.5))
```

## Завдання 6

```python
def is_a_triangle(a, b, c):
    """
    Перевіряє, чи можуть сторони a, b, c утворити трикутник.
    """
    return (a > 0 and b > 0 and c > 0 and
            a + b > c and
            a + c > b and
            b + c > a)

print("--- Запуск тестів ---")

print("Трикутник (3, 4, 5):", is_a_triangle(3, 4, 5))
print("Трикутник (1, 1, 1):", is_a_triangle(1, 1, 1))
print("Трикутник (10, 10, 1):", is_a_triangle(10, 10, 1))

print("\nНеможливий (1, 2, 3) - вироджений:", is_a_triangle(1, 2, 3)) # 1+2 не > 3
print("Неможливий (1, 2, 5):", is_a_triangle(1, 2, 5))         # 1+2 не > 5
print("Неможливий (5, 1, 2):", is_a_triangle(5, 1, 2))         # 1+2 не > 5
print("Неможливий (0, 1, 1) - нульова сторона:", is_a_triangle(0, 1, 1))
print("Неможливий (-1, 1, 1) - від'ємна сторона:", is_a_triangle(-1, 1, 1))
```

## Завдання 7

```python
def is_a_triangle(a, b, c):
    """
    Перевіряє, чи можуть сторони a, b, c утворити трикутник.
    """
    return (a > 0 and b > 0 and c > 0 and
            a + b > c and
            a + c > b and
            b + c > a)

def is_a_right_triangle(a, b, c):
    """
    Перевіряє, чи є трикутник зі сторонами a, b, c прямокутним.
    """
    if not is_a_triangle(a, b, c):
        return False  
    
    sides = sorted([a, b, c])
    c1 = sides[0]    
    c2 = sides[1]   
    hypot = sides[2] 
    
    epsilon = 1e-9 
    
    if abs((c1**2 + c2**2) - hypot**2) < epsilon:
        return True
    else:
        return False

print("--- Запуск тестів ---")

print("Прямокутний (5, 3, 4):", is_a_right_triangle(5, 3, 4))
print("Прямокутний (3, 4, 5):", is_a_right_triangle(3, 4, 5))

print("Прямокутний (13, 12, 5):", is_a_right_triangle(13, 12, 5))

print("Не прямокутний (1, 1, 1):", is_a_right_triangle(1, 1, 1))

print("Не трикутник (1, 2, 3):", is_a_right_triangle(1, 2, 3))

print("Не трикутник (0, 4, 5):", is_a_right_triangle(0, 4, 5))

print("Прямокутний (0.5, 0.4, 0.3):", is_a_right_triangle(0.5, 0.4, 0.3))
```
