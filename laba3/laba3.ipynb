```python
import math as m

x = 1
mu = 0
sigma = 1

f = (1 / (sigma * m.sqrt(2 * m.pi))) * m.exp(- ((x - mu) ** 2) / (2 * sigma ** 2))

print(f"Значення функції Гауса при x={x}, \u03bc={mu}, \u03c3={sigma}: {f:.5f}")
```

```python
import math

def gauss(x, mu=0, sigma=1):
    return (1 / (sigma * m.sqrt(2 * m.pi))) * m.exp(- ((x - mu) ** 2) / (2 * sigma ** 2))

# print(gauss(1))

# print(gauss(1, mu=2, sigma=2))

x = int(input("Уведіть x: "))

print(f"x= {x}: {gauss(x, mu=2, sigma=2):.5f}")
```

## Завдання 5

```python
# Програма для обчислення кількості секунд у заданій кількості годин.

hours = 2  # Кількість годин для перетворення
seconds_per_hour = 3600  # Кількість секунд в одній годині

# Розрахунок та виведення результату
total_seconds = hours * seconds_per_hour

print(f"Кількість годин: {hours}")
print(f"Кількість секунд у {hours} годинах: {total_seconds}")
print("До побачення!")
```

## Завдання 6

```python
a = float(input("Введіть перше число (a): "))

b = float(input("Введіть друге число (b): "))

print("\n--- Результати ---")

print(f"{a} + {b} = {a + b}")

print(f"{a} - {b} = {a - b}")

print(f"{a} * {b} = {a * b}")

print(f"{a} / {b} = {a / b}")

print("\nThat's all, folks!")
```
