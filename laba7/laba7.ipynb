## Завдання 1

```python
def filter_numbers():
    source_list = [10, 5, 25, 3, 8, 15, 1, 42, 7, 12]

    data_tuple = tuple(source_list)
    
    print(f"Наявний кортеж: {data_tuple}")

    try:
        n = int(input("Введіть число n (порогове значення): "))

        result = [num for num in data_tuple if num < n]

        print(f"Числа, які менші за {n}: {result}")

    except ValueError:
        print("Помилка: Будь ласка, введіть ціле число.")

if __name__ == "__main__":
    filter_numbers()
```

## Завдання 2

```python
def join_tuple_strings():
    words_tuple = ("Python", "Java", "C++")
    
    print(f"Початковий кортеж: {words_tuple}")

    result_string = ", ".join(words_tuple)

    print(f"З'єднаний рядок: {result_string}")
    
if __name__ == "__main__":
    join_tuple_strings()
```

## Завдання 3

```python
def library_system():
    library = {
        "Кобзар": {
            "author": "Тарас Шевченко",
            "year": 1840,
            "pages": 115
        },
        "1984": {
            "author": "Джордж Орвелл",
            "year": 1949,
            "pages": 328
        },
        "Гаррі Поттер": {
            "author": "Дж. К. Роулінг",
            "year": 1997,
            "pages": 336
        },
        "Тіні забутих предків": {
            "author": "Михайло Коцюбинський",
            "year": 1911,
            "pages": 250
        }
    }

    print("Доступні книги:", ", ".join(library.keys()))
    book_name = input("\nВведіть назву книги, яку хочете знайти: ")

    if book_name in library:
        book_info = library[book_name]
        
        print(f"\n--- Інформація про книгу '{book_name}' ---")
        print(f"Автор: {book_info['author']}")
        print(f"Рік видання: {book_info['year']}")
        print(f"Кількість сторінок: {book_info['pages']}")
    else:
        print(f"\nНа жаль, книги '{book_name}' немає у нашій бібліотеці.")

if __name__ == "__main__":
    library_system()
```

## Завдання 4

```python
def student_database():
    students = {
        "Коваленко": ("Олександр", 19, 2, 88.5),
        "Бондаренко": ("Марія", 20, 3, 92.0),
        "Шевченко": ("Андрій", 18, 1, 75.4),
        "Мельник": ("Олена", 21, 4, 95.0)
    }

    print("У базі є такі студенти:", ", ".join(students.keys()))

    search_surname = input("\nВведіть прізвище студента: ").strip().title()

    if search_surname in students:
        info = students[search_surname]
        
        print(f"\n--- Інформація про студента {search_surname} ---")
        print(f"Ім'я: {info[0]}")
        print(f"Вік: {info[1]} років")
        print(f"Курс: {info[2]}")
        print(f"Середній бал: {info[3]}")
    else:
        print(f"\nСтудента з прізвищем '{search_surname}' не знайдено.")

if __name__ == "__main__":
    student_database()
```

## Завдання 5

```python
phone_book = {
    "Олександр": ["050-111-22-33", "097-123-45-67"],
    "Марина": ["063-999-88-77"],
    "Робота": ["044-555-00-11", "044-555-00-12", "0-800-200-100"]
}

def add_phone_number(contact_name, new_number):
    """
    Функція додає новий номер до існуючого контакту 
    або створює нового, якщо контакту немає.
    """
    if contact_name in phone_book:
        phone_book[contact_name].append(new_number)
        print(f" Номер {new_number} успішно додано до контакту '{contact_name}'.")
    else:
        phone_book[contact_name] = [new_number]
        print(f" Створено новий контакт '{contact_name}' з номером {new_number}.")

add_phone_number("Марина", "095-000-01-01")

add_phone_number("Служба таксі", "303")

print("\n---  ОНОВЛЕНА ТЕЛЕФОННА КНИГА ---")

for name, numbers in phone_book.items():
    formatted_numbers = ", ".join(numbers)
    print(f" {name}: {formatted_numbers}")
```
